```python
from id3 import Id3Estimator, export_text,export_graphviz
import numpy as np
import graphviz
from graphviz import Source
import csv


d=[]
col=[]
op=[]
with open('unemployment.csv','r')as f:
    data = csv.reader(f)
    for row in data:
        d.append(row)


feature_names = d[0][1:10]
for i in range(1,453):
    col.append(d[i][1:10])
    op.append(d[i][1])
X = col

y = op

clf = Id3Estimator()
clf.fit(X, y, check_input=True)

print(export_text(clf.tree_, feature_names))

export_graphviz(clf.tree_, "ot.dot", feature_names)
s = Source.from_file("ot.dot")
s.view()
```
